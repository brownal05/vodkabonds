##### CAPM model (E (Ra) = Rf + Ba [ E (Rm) - Rf])


```python
def linreg(x,y):
    x = sm.add_constant(x)
    model = sm.OLS(y,x).fit()
    x = x[:,1]
    return model.params[0], model.params[1]

X = diff_ov.values
Y = diff_rs.values
alpha, beta = linreg(X,Y)    
```

```python
print("Russian 10 yr beta coeff")
print('alpha ' + str(alpha))
print('beta '+ str(beta))
```

```python
X2 = np.linspace(X.min(), X.max(), 100)
Y_hat = X2 * beta + alpha
plt.scatter(X, Y, alpha=0.3)
plt.xlabel("UST10")
plt.ylabel("Russian 10 Yr Benchmark")
plt.plot(X2, Y_hat, 'r', alpha=0.9)
plt.show()
```
